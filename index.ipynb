```python
def get_album_links(sp):
    raw_album_links = sp.find(id='listAlbum').findAll('a')

    album_links = []
    base = 'https://www.azlyrics.com/t'
    for tag in raw_album_links:
        if tag.get('href'):
            link =  base+tag.get('href')[2:]
            album_links.append(link)
    return album_links
```
